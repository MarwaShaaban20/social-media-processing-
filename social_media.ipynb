```python
#encode gender
df['Encoded Sex'] = df['Gender'].map({'Male':1, 'Female':0})
df['Encoded Affects_Academic_Performance'] = df['Affects_Academic_Performance'].map({'Yes':1, 'No':0})
```

```python
correlation = df['Avg_Daily_Usage_Hours'].corr(df['Mental_Health_Score'])
print("Correlation:", correlation)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lmplot(data=df, x='Avg_Daily_Usage_Hours', y='Mental_Health_Score')
plt.title(f"Relationship (Correlation = {correlation:.2f})")
plt.show()
```

```python

def classify_addiction(score):
    if score <= 3:
        return 'Low'
    elif score <= 6:
        return 'Moderate'
    else:
        return 'High'


df['Addiction_Level'] = df['Addicted_Score'].apply(classify_addiction)
df
```

```python
df['Addiction_Level'].value_counts()
```

```python
def classify_mental_health(score):# كل م زاد الرقم اتحسنت الحاله النفسيه
    if score <= 3:
        return 'Poor'
    elif score <= 6:
        return 'Average'
    else:
        return 'Good'


df['Mental_Health_Level'] = df['Mental_Health_Score'].apply(classify_mental_health)
df
```

```python
# حساب متوسط عدد ساعات الاستخدام لكل دولة
top_countries = df.groupby('Country')['Avg_Daily_Usage_Hours'].mean().sort_values(ascending=False).head(5)

# عرض النتائج
print(top_countries)
```

```python
import matplotlib.pyplot as plt

top_countries.plot(kind='bar', color='teal')
plt.title("Top 5 Countries by Avg. Daily Social Media Usage")
plt.xlabel("Country")
plt.ylabel("Average Daily Usage (Hours)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
